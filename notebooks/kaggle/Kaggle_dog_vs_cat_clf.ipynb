```python
# device = 'mps' if torch.backends.mps.is_available else 'cpu'
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
train_path = '/content/train/train'
test_path = '/content/test'

# train_path = '/kaggle/input/dog-vs-cat-classification/train/train'
# test_path = '/kaggle/input/dog-vs-cat-classification/test'

# train_path = '../../data/dog-vs-cat-classification/train/train/'
# test_path = '../../data//dog-vs-cat-classification/test'

transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])

train_image_folder = ImageFolder(train_path, transform=transform)
test_image_folder = ImageFolder(test_path, transform=transform)

batch_size = 32
train_loader = DataLoader(train_image_folder, shuffle=True, batch_size=batch_size)
test_loader = DataLoader(test_image_folder, shuffle=False)
```

```python
fig, ax = plt.subplots(ncols=8, nrows=2, figsize=(16, 4))
n = 0
for i in range(2):
    for j in range(8):
        if i == 0:
            ax[i][j].imshow(train_image_folder[n][0].permute(1, 2, 0))
        else:
            ax[i][j].imshow(train_image_folder[len(train_image_folder) - 1 - n][0].permute(1, 2, 0))
        ax[i][j].axis(False)
        n += 1
plt.show()
```

```python
class EfficientnetB4(nn.Module):
    def __init__(self, num_classes=1):
        super(EfficientnetB4, self).__init__()
        self.efficientnet = efficientnet_b4(weights=None)
        self.efficientnet.classifier[1] = nn.Linear(1792, num_classes)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.efficientnet(x)
        x = self.sigmoid(x)
        return x

model = EfficientnetB4()
model.to(device)

weights = torch.load('/content/efficientnet_b4.pth')
model.load_state_dict(weights)
```

```python
def fit(model, loss_func, opt, train_loader, test_loader, epochs, scheduler):
    train_loss_hist = []
    test_loss_hist = []
    accuracy_hist = []

    for epoch in tqdm(range(epochs)):
        model.train()
        avg_loss = 0
        avg_accuracy = 0
        for xb, yb in tqdm(train_loader):
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad()
            logits = model(xb).squeeze()
            loss = loss_func(logits, yb.float())
            loss.backward()
            opt.step()
            avg_loss += loss.item()

            y_true = yb.cpu().numpy()
            y_pred = (logits >= 0.5).int().cpu().numpy()
            avg_accuracy += accuracy_score(y_true, y_pred)

        train_loss_hist.append(avg_loss / len(train_loader))
        accuracy_hist.append(avg_accuracy / len(train_loader))
        scheduler.step()

#         model.eval()
#         with torch.no_grad():
#             model.eval()
#             y_pred = []
#             y_true = []
#             avg_loss = 0
#             for x, y in test_loader:
#                 x, y = x.to(device), y.to(device)
#                 logit = model(x)
#                 loss = loss_func(logit[0], y.float())
#                 avg_loss += loss.item()
#                 pred = (logit[0] >= 0.5).float()
#                 y_pred.append(pred.item())
#                 y_true.append(y.item())
#             test_loss_hist.append(avg_loss / len(test_loader))
#             accuracy_hist.append(accuracy_score(y_true, y_pred))

        train_loss = round(train_loss_hist[-1], 5)
#         test_loss = round(test_loss_hist[-1], 5)
        accuracy = round(accuracy_hist[-1], 5)
        lr = scheduler.get_last_lr()[0]

        print(f"Epoch {epoch+1}'st: train_loss = {train_loss}, train_accuracy = {accuracy}, lr = {lr}")
#         print(f"Epoch {epoch+1}'st: train_loss = {train_loss}, test_loss = {test_loss}, accuracy = {accuracy}, lr = {lr}")

    return train_loss_hist, test_loss_hist, accuracy_hist
```

```python
epochs = 25
loss_func = nn.BCELoss()
opt = optim.Adam(model.parameters(), lr=0.001)
scheduler = lr_scheduler.MultiStepLR(opt, milestones=[10, 15, 20], gamma=0.5)

hist = fit(model, loss_func, opt, train_loader, test_loader, epochs, scheduler)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 4))
x = range(epochs)

sns.lineplot(x=x, y=hist[0], label='train_loss', ax=ax[0])
# sns.lineplot(x=x, y=hist[1], label='test_loss', ax=ax[0])
sns.lineplot(x=x, y=hist[2], label='train_accuracy', ax=ax[1])

ax[0].set_xticks(range(epochs+1))
ax[1].set_xticks(range(epochs+1))
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(ncols=8, figsize=(15, 5))

iterator = iter(train_loader)
model.eval()
with torch.no_grad():
    for i in range(8):
        input = next(iterator)
        x = input[0][0:1].to(device)
        y = input[1][0:1].to(device)
        output = model(x)
        logit = torch.sigmoid(output)
        y_pred = (logit[0] >= 0.5).int()
        image = x[0].permute(1, 2, 0).cpu().numpy()
        ax[i].imshow(image)
        ax[i].axis(False)
        ax[i].set_title(f'y_true: {y.item()} \ny_pred: {y_pred.item()}')

plt.show()
```

```python
def eval_model(model, test_loader):
    y_pred = []
    model.eval()
    with torch.no_grad():
        for x, y in tqdm(test_loader):
            x = x.to(device)
            logit = model(x)
            pred = (logit[0] >= 0.5).int()
            y_pred.append(pred.item())
    return y_pred
```

```python
y_pred = eval_model(model, test_loader)
unique_values, counts = np.unique(y_pred, return_counts=True)
for value, count in zip(unique_values, counts):
    print(f"Value: {value}, Count: {count}")
```
